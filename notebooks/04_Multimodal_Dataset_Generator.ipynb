```python
from docx import Document
import os

def generate_document(document_type, data, file_number):

    doc = Document()

    doc.add_heading(document_type, level=1)

    for key, value in data.items():
        doc.add_paragraph(f"{key}: {value}")

    folder = document_type.replace(" ", "_")

    os.makedirs(folder, exist_ok=True)

    filename = f"{folder}/{folder}_{file_number:03d}.docx"

    doc.save(filename)

    return filename
```

```python
def generate_bulk_documents(document_type, count):

    generated_files = []

    for i in range(1, count + 1):

        if document_type == "Salary Slip":

            data = create_salary_data()

            department = "HR"

            classification = "Confidential"

            risk = "High"

        elif document_type == "Invoice":

            data = create_invoice_data()

            department = "Finance"

            classification = "Internal"

            risk = "Medium"

        elif document_type == "Contract":

            data = create_contract_data()

            department = "Legal"

            classification = "Restricted"

            risk = "Critical"

        else:

            continue

        filename = generate_document(
            document_type,
            data,
            i
        )

        generated_files.append(filename)

        metadata.append({

            "File":filename,

            "DocumentType":document_type,

            "Department":department,

            "Classification":classification,

            "Risk":risk

        })

    return generated_files
```

```python
from reportlab.platypus import SimpleDocTemplate, Paragraph
from reportlab.lib.styles import getSampleStyleSheet

from PIL import Image, ImageDraw
import os
```

```python
def export_txt(data, filename):

    with open(filename, "w") as f:

        for key, value in data.items():

            f.write(f"{key}: {value}\n")

    print(filename, "Created")
```

```python
def export_pdf(data, filename):

    doc = SimpleDocTemplate(filename)

    styles = getSampleStyleSheet()

    story = []

    story.append(Paragraph("<b>Generated Document</b>", styles["Heading1"]))

    for key, value in data.items():

        story.append(
            Paragraph(f"<b>{key}</b>: {value}", styles["Normal"])
        )

    doc.build(story)

    print(filename, "Created")
```

```python
def export_png(data, filename):

    image = Image.new("RGB", (800, 900), "white")

    draw = ImageDraw.Draw(image)

    y = 30

    draw.text((20, 10), "Generated Document", fill="black")

    for key, value in data.items():

        draw.text((20, y), f"{key}: {value}", fill="black")

        y += 30

    image.save(filename)

    print(filename, "Created")
```

```python
def export_document(data, document_type, number):

    folder = document_type.replace(" ", "_")

    os.makedirs(folder, exist_ok=True)

    base = f"{folder}/{folder}_{number:03d}"

    # TXT
    export_txt(data, base + ".txt")

    # DOCX
    generate_document(document_type, data, number)

    # PDF
    export_pdf(data, base + ".pdf")

    # PNG
    export_png(data, base + ".png")
```

```python
from docx import Document
from docx.shared import Pt

def generate_bank_statement(data, number):

    doc = Document()

    doc.add_heading(
        data["Bank Name"],
        level=1
    )

    doc.add_heading(
        "BANK STATEMENT",
        level=2
    )

    doc.add_paragraph(
        f"Account Holder : {data['Account Holder']}"
    )

    doc.add_paragraph(
        f"Account Number : {data['Account Number']}"
    )

    doc.add_paragraph(
        f"Branch : {data['Branch']}"
    )

    doc.add_paragraph(
        f"Current Balance : Rs.{data['Current Balance']}"
    )

    doc.add_heading(
        "Transactions",
        level=2
    )

    table = doc.add_table(
        rows=1,
        cols=3
    )

    table.style = "Table Grid"

    hdr = table.rows[0].cells

    hdr[0].text = "Date"

    hdr[1].text = "Description"

    hdr[2].text = "Amount"

    for row in data["Transactions"]:

        cells = table.add_row().cells

        cells[0].text = str(row[0])

        cells[1].text = row[1]

        cells[2].text = str(row[2])

    folder = "Bank_Statement"

    import os

    os.makedirs(folder,exist_ok=True)

    filename = f"{folder}/Bank_Statement_{number:03d}.docx"

    doc.save(filename)

    print(filename,"Created")
```

```python
classification_registry = {

    "Salary Slip":("HR","Confidential","High"),

    "Invoice":("Finance","Internal","Medium"),

    "Contract":("Legal","Restricted","Critical"),

    "Bank Statement":("Finance","Confidential","High"),

    "Leave Form":("HR","Internal","Low"),

    "HR Policy":("HR","Internal","Low"),

    "Meeting Minutes":("Operations","Internal","Medium"),

    "NDA":("Legal","Restricted","Critical"),

    "Tax Report":("Finance","Confidential","High"),

    "Firewall Configuration":("IT","Restricted","Critical"),

    "Source Code":("IT","Restricted","Critical"),

    "Marketing Brochure":("Marketing","Public","Low")

}
```

```python
import pandas as pd

def generate_complete_dataset(documents_per_type=10):

    metadata = []

    for document_type, generator in document_generators.items():

        template = template_registry[document_type]

        department, classification, risk = classification_registry[document_type]

        for i in range(1, documents_per_type + 1):

            data = generator()

            generate_document_from_template(

                template,

                data,

                i

            )

            folder = document_type.replace(" ", "_")

            base = f"{folder}/{folder}_{i:03d}"

            for fmt in ["txt","docx","pdf","png"]:

                metadata.append({

                    "File Name":f"{base}.{fmt}",

                    "Document Type":document_type,

                    "Department":department,

                    "Classification":classification,

                    "Risk":risk,

                    "Format":fmt.upper()

                })

    df = pd.DataFrame(metadata)

    df.to_csv(

        "dataset_metadata.csv",

        index=False

    )

    print("=================================")

    print("Dataset Generation Completed")

    print("Metadata Saved")

    print("=================================")

    return df
```

```python
def generate_document_id(number):

    return f"DOC{number:06d}"

print(generate_document_id(1))
```

```python
import pandas as pd
import random

def generate_complete_dataset(documents_per_type=100):

    metadata = []

    document_counter = 1

    for document_type, generator in document_generators.items():

        template = template_registry[document_type]

        department, classification, risk = classification_registry[document_type]

        industry = industry_registry[document_type]

        entities = entity_registry[document_type]

        print(f"\nGenerating {document_type}...")

        for i in range(1, documents_per_type + 1):

            data = generator()

            generate_document_from_template(

                template,

                data,

                i

            )

            folder = document_type.replace(" ", "_")

            base = f"{folder}/{folder}_{i:03d}"

            for fmt in ["txt", "docx", "pdf", "png"]:

                metadata.append({

                    "Document ID": generate_document_id(document_counter),

                    "File Name": f"{base}.{fmt}",

                    "Document Type": document_type,

                    "Department": department,

                    "Industry": industry,

                    "Classification": classification,

                    "Risk": risk,

                    "Sensitive Entities": ", ".join(entities),

                    "Leakage Score": random.randint(60,100),

                    "Format": fmt.upper()

                })

                document_counter += 1

    df = pd.DataFrame(metadata)

    df.to_csv(

        "dataset_metadata.csv",

        index=False

    )

    print("\n======================================")

    print("✅ Custom Dataset Generated Successfully")

    print(f"📄 Total Metadata Records : {len(df)}")

    print("💾 dataset_metadata.csv Saved")

    print("======================================")

    return df
```
