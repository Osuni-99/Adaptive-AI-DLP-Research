```python
def generate_employee():

    employee = {

        "EmployeeID": fake.random_int(1000,9999),

        "Name": fake.name(),

        "Department": random.choice([
            "HR",
            "Finance",
            "IT",
            "Marketing",
            "Legal"
        ]),

        "Salary": random.randint(50000,300000),

        "NIC": fake.random_number(digits=12),

        "Email": fake.email()

    }

    return employee
```

```python
def generate_employee():

    classification = random.choice([
        "Public",
        "Internal",
        "Confidential",
        "Restricted"
    ])

    risk = {
        "Public":"Low",
        "Internal":"Medium",
        "Confidential":"High",
        "Restricted":"Critical"
    }

    score = {
        "Public":random.randint(1,25),
        "Internal":random.randint(26,50),
        "Confidential":random.randint(51,80),
        "Restricted":random.randint(81,100)
    }

    employee = {

        "EmployeeID":fake.random_int(1000,9999),

        "Name":fake.name(),

        "Department":random.choice([
            "HR",
            "Finance",
            "IT",
            "Marketing",
            "Legal"
        ]),

        "Salary":random.randint(50000,300000),

        "NIC":fake.random_number(digits=12),

        "BankAccount":fake.random_number(digits=10),

        "Email":fake.email(),

        "DocumentType":"Employee Record",

        "Classification":classification,

        "RiskLevel":risk[classification],

        "LeakageScore":score[classification]

    }

    return employee
```

```python
employees=[]

for i in range(500):

    employees.append(generate_employee())

df=pd.DataFrame(employees)

df.head()
```

```python
df.to_csv("SME_DLP_Dataset.csv",index=False)

print("Dataset Created Successfully")
```
