```python
department_rules = [

    ["HR","Salary",3],

    ["HR","Employee ID",2],

    ["Finance","Bank Account",4],

    ["Finance","Invoice",3],

    ["Finance","Tax",2],

    ["IT","Password",5],

    ["IT","API Key",5],

    ["IT","Firewall",4],

    ["Legal","Contract",4],

    ["Legal","Agreement",3],

    ["Marketing","Brochure",1]

]
```

```python
import pandas as pd

risk_df = pd.read_csv("risk_rules.csv")
department_df = pd.read_csv("department_rules.csv")

print("Rules Loaded Successfully")
```

```python
sample_document = {

    "Department":"Finance",

    "DocumentType":"Salary Slip",

    "Text":"""

    Employee Name : Kasun Silva

    Salary : LKR 250000

    Bank Account : 456789123

    Password : Admin123

    """

}

sample_document
```

```python
text = sample_document["Text"].lower()

detected_features = []

for _, row in risk_df.iterrows():

    if row["Feature"].lower() in text:

        detected_features.append(row["Feature"])

print(detected_features)
```

```python
rule_score = 0

for feature in detected_features:

    weight = risk_df.loc[
        risk_df["Feature"] == feature,
        "Weight"
    ].values[0]

    rule_score += weight

print("Rule Score :", rule_score)
```

```python
department = sample_document["Department"]

extra_weight = 0

for feature in detected_features:

    rows = department_df[
        (department_df["Department"] == department) &
        (department_df["Feature"] == feature)
    ]

    if not rows.empty:
        extra_weight += rows["ExtraWeight"].values[0]

print("Department Weight :", extra_weight)
```

```python
final_score = rule_score + extra_weight

print("Final Risk Score :", final_score)
```

```python
if final_score >= 30:

    risk = "Critical"

elif final_score >= 20:

    risk = "High"

elif final_score >= 10:

    risk = "Medium"

else:

    risk = "Low"

print("Risk Level :", risk)
```
