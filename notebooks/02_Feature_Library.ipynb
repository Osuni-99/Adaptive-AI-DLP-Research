```python
import re
import pandas as pd

print("Feature Library Started")
```

```python
SENSITIVE_FEATURES = {

    "Personal Information":[

        "NIC",
        "National Identity Card",
        "Passport",
        "Driving License",
        "Date of Birth",
        "Phone Number",
        "Email Address"

    ],

    "Financial Information":[

        "Bank Account",
        "Credit Card",
        "Debit Card",
        "Salary",
        "Invoice",
        "Tax",
        "Payment"

    ],

    "Business Information":[

        "Confidential",
        "Agreement",
        "Contract",
        "Business Plan",
        "Customer List",
        "Source Code"

    ],

    "Security Information":[

        "Password",
        "Firewall",
        "Encryption Key",
        "VPN",
        "Administrator",
        "Access Token"

    ]

}

print(SENSITIVE_FEATURES)
```

```python
rows=[]

for category,keywords in SENSITIVE_FEATURES.items():

    for keyword in keywords:

        rows.append({

            "Category":category,
            "Keyword":keyword

        })

feature_df=pd.DataFrame(rows)

feature_df.head(20)
```

```python
feature_df.to_csv("Sensitive_Feature_Library.csv",index=False)

print("Feature Library Saved")
```

```python
sample_text="""

Employee Name : Kasun Silva

NIC : 200012345678

Salary : LKR 250000

Bank Account : 45678912345

Password : Admin@123

"""

detected=[]

for category,keywords in SENSITIVE_FEATURES.items():

    for keyword in keywords:

        if keyword.lower() in sample_text.lower():

            detected.append([category,keyword])

pd.DataFrame(detected,columns=["Category","Detected"])
```
