```python
class DatasetManager:

    def __init__(self):

        self.datasets={}

    def register(self,name,dataframe,category):

        self.datasets[name]={

            "data":dataframe,

            "category":category

        }

    def summary(self):

        rows=[]

        for name,info in self.datasets.items():

            df=info["data"]

            rows.append({

                "Dataset":name,

                "Category":info["category"],

                "Rows":len(df),

                "Columns":len(df.columns)

            })

        return pd.DataFrame(rows)
```
